# 대학교 주소 표시

```python
import requests
from urllib.parse import urlparse
import pandas as pd
```

```python
key_fd = open('roadapikey.txt', mode='r')
roadapi_key = key_fd.read(100)
key_fd.close()
```

```python
univs = ['서울대학교', '연세대학교', '고려대학교', '성균관대학교', '서강대학교', '중앙대학교']
road_url = 'https://www.juso.go.kr/addrlink/addrLinkApi.do'
road_addr_list = []
for univ in univs:
    url = f'{road_url}?confmKey={roadapi_key}&currentPage=1&coutPerPage=10&keyword={univ}&resultType=json'
    results = requests.get(urlparse(url).geturl()).json()
    results = results['results']['juso']
results
```
