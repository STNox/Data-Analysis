# 서울시 강서구 커피 매장

```python
import requests
from urllib.parse import urlparse
import pandas as pd
import folium
```

```python
df = pd.read_csv('storeList.csv', sep='|')
df = df[df['상권업종중분류명'] == '커피점/카페']
df = df.reset_index()
cafes = df.loc[df['도로명주소'].str.contains('강서구')]
starbucks_df = cafes.loc[cafes['상호명'].str.contains('스타벅스' or 'STARBUCKS')]
starbucks_df['상호명'] = '스타벅스'
ediya_df = cafes.loc[cafes['상호명'].str.contains('이디야' or 'EDIYA')]
ediya_df['상호명'] = '이디야'
paik_df = cafes.loc[cafes['상호명'].str.contains('빽다방' or 'PAIK')]
paik_df['상호명'] = '빽다방'
twosome_df = cafes.loc[cafes['상호명'].str.contains('투썸플레이스' or 'TWOSOME')]
twosome_df['상호명'] = '투썸플레이스'
coffeebean_df = cafes.loc[cafes['상호명'].str.contains('커피빈' or 'COFFEEBEAN')]
osome_df = cafes.loc[cafes['상호명'].str.contains('투썸플레이스' or 'TWOSOME')]
twosome_df['상호명'] = '투썸플레이스'
coffeebean_df = cafes.loc[cafes['상호명'].str.contains('커피빈' or 'COFFEEBEAN')]
coffeebean_df['상호명'] = '커피빈'
cafes_df = pd.concat([starbucks_df, ediya_df, paik_df, twosome_df, coffeebean_df])
cafes_df
```

```python
key_fd = open('roadapikey.txt', mode='r')
roadapi_key = key_fd.read(100)
key_fd.close()
```

```python
colors_dict = {'스타벅스': 'red', '이디야': 'green', '빽다방': 'blue', '투썸플레이스': 'gray', '커피빈': 'purple'}
mapping = folium.Map(location=[cafes_df['위도'].mean(), cafes_df['경도'].mean()], zoom_start=14)
for i in cafes_df.index:
    folium.Marker(
        location=[cafes_df['위도'][i], cafes_df['경도'][i]],
        popup=cafes_df['상호명'][i],
        icon=folium.Icon(color=colors_dict[cafes_df['상호명'][i]])
    ).add_to(mapping)
mapping
```
