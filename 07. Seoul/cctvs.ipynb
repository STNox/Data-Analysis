```python
import pandas as pd
```

```python
pop_Seoul = pd.read_csv('./data/population_Seoul.csv', delimiter='\t', header=2)[['자치구', '계', '계.1', '계.2', '65세이상고령자']]
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={pop_Seoul.columns[0] : '구별', pop_Seoul.columns[1] : '인구수', pop_Seoul.columns[2] : '한국인', pop_Seoul.columns[3] : '외국인', pop_Seoul.columns[4] : '고령자'}, inplace=True)
pop_Seoul.head()
```

```python
import numpy as np
```

```python
""" for i in range(len(CCTV_Seoul['소계'])):
    CCTV_Seoul['소계'][i] = CCTV_Seoul['소계'][i].replace(',', '')
CCTV_Seoul['소계'] = pd.to_numeric(CCTV_Seoul['소계']) """
CCTV_Seoul = CCTV_Seoul.fillna(0)
```

```python
for i in range(len(CCTV_Seoul['소계'])):
    if type(CCTV_Seoul['2011년 이전'][i]) == str:
        CCTV_Seoul['2011년 이전'][i] = CCTV_Seoul['2011년 이전'][i].replace(',', '')
CCTV_Seoul['2011년 이전'] = pd.to_numeric(CCTV_Seoul['2011년 이전'])
```

```python
for i in range(len(CCTV_Seoul['소계'])):
    if type(CCTV_Seoul['2014년'][i]) == str:
        CCTV_Seoul['2014년'][i] = CCTV_Seoul['2014년'][i].replace(',', '')
CCTV_Seoul['2014년'] = pd.to_numeric(CCTV_Seoul['2014년'])
```

```python
CCTV_Seoul.sort_values(by='소계', ascending=True).head()
```

```python
CCTV_Seoul.sort_values(by='소계', ascending=False).head()
```

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2018년'] + CCTV_Seoul['2017년'] + CCTV_Seoul['2016년']) / (CCTV_Seoul['2015년'] + CCTV_Seoul['2014년'] + CCTV_Seoul['2013년'] + CCTV_Seoul['2012년'] + CCTV_Seoul['2011년 이전']) * 100
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head()
```

```python
for i in range(len(pop_Seoul['구별'])):
    if type(pop_Seoul['인구수'][i]) == str:
        pop_Seoul['인구수'][i] = pop_Seoul['인구수'][i].replace(',', '')
pop_Seoul['인구수'] = pd.to_numeric(pop_Seoul['인구수'])
for i in range(len(pop_Seoul['구별'])):
    if type(pop_Seoul['한국인'][i]) == str:
        pop_Seoul['한국인'][i] = pop_Seoul['한국인'][i].replace(',', '')
pop_Seoul['한국인'] = pd.to_numeric(pop_Seoul['한국인'])
for i in range(len(pop_Seoul['구별'])):
    if type(pop_Seoul['외국인'][i]) == str:
        pop_Seoul['외국인'][i] = pop_Seoul['외국인'][i].replace(',', '')
pop_Seoul['외국인'] = pd.to_numeric(pop_Seoul['외국인'])
for i in range(len(pop_Seoul['구별'])):
    if type(pop_Seoul['고령자'][i]) == str:
        pop_Seoul['고령자'][i] = pop_Seoul['고령자'][i].replace(',', '')
pop_Seoul['고령자'] = pd.to_numeric(pop_Seoul['고령자'])
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))

df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14, 10))
plt.scatter(data_result['인구수'], data_result['소계'], c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n] * 1.02, df_sort['소계'][n] * 0.98, df_sort.index[n], fontsize=15)

plt.xlabel('인구수')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```
