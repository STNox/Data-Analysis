# 데이터 시각화

### 시각화 패키지 맷플롯리브

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

### 라인 플롯

```python
plt.plot([1, 4, 9, 16])         # x 좌표는 [0, 1, 2, 3, ...]이 default
plt.title('Line Plot')
plt.show()
```

```python
plt.plot([1, 4, 9, 16])
plt.title('Line Plot')
plt.grid(True)
plt.show()
```

```python
plt.plot([1, 4, 9, 16])
plt.title('Line Plot')
plt.grid(True)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
plt.title('x ticks')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.grid(True)
plt.show()
```

```python
plt.title('x 틱스')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.grid(True)
plt.show()
```

### 한글 사용

```python
import matplotlib.font_manager

matplotlib.font_manager._rebuild()
sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist if f.name.startswith("Malgun")])
```

```python
# 폰트 설정
mpl.rc('font', family='Malgun Gothic')
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
plt.title('한글 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel('엑스축 라벨')
plt.ylabel('와이축 라벨')
plt.show()
```

```python
font1 = {'family': 'D2Coding', 'size': 24, 
         'color':  'black'}
font2 = {'family': 'Malgun Gothic', 'size': 18, 'weight': 'bold',
         'color':  'darkred'}
font3 = {'family': 'Batang', 'size': 12, 'weight': 'light',
         'color':  'blue'}

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title('한글 제목', fontdict=font1)
plt.xlabel('엑스 축', fontdict=font2)
plt.ylabel('와이 축', fontdict=font3)
plt.show()
```
