### Matplotlib 그래프 영역 채우기

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-label'); plt.ylabel('Y-label')
plt.fill_between(x[1:3], y[1:3], alpha=0.2) # alpha: 투명도
plt.axhline(0); plt.axvline(0) # 수직, 수평선
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-label'); plt.ylabel('Y-label')
plt.fill_betweenx(y[1:3], x[1:3], color='pink', alpha=0.5)
plt.axhline(0); plt.axvline(0)
plt.show()
```

```python
x = np.arange(1, 4, 0.01)
y1 = x ** 2
y2 = x ** 2 / 2

plt.title('두 그래프 사이 영역 채우기')
plt.plot(x, y1); plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.fill_between(x[100:200], y1[100:200], y2[100:200], color='lightgray', alpha=0.5)
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1); plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.fill([1.9, 1.9, 3.1, 3.1], [2, 5, 11, 8], color='lightgray', alpha=0.5)

plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 256) # (a, b, n) a와 b 사이의 256개 값을 제공
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='dodgerblue', label='sine'); plt.plot(x, y2, color='darkred', label='cosine')
plt.axhline(0, 0, 1, color='gray', linestyle='-', linewidth='1')
plt.axvline(0, 0, 1, color='gray', linestyle='-', linewidth='1')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], labels=['0', 'π/2', 'π', '3π/2', '2π'])
plt.yticks([-1, 0, 1], labels=['-1', '0', '1'])
plt.grid(True, color='gray', alpha=0.5, linestyle='--')
plt.legend(loc='upperright')
plt.show()
```

### 막대 그래프 그리기

```python
x = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
```

```python
plt.bar(x, values, width=0.6, align='edge', color='dodgerblue', edgecolor='black', linewidth=3, tick_label=years, log=True)
plt.show()
```

### 수평 막대 그래프 그리기

### 산점도 그리기

```python
np.random.seed(2020)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()
```

```python
np.random.seed(2020)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2        # 값이 클수록 마커의 크기도 증가

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

### 히스토그램 여러 개 그리기

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step', label='a')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled', label='b')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step', label='c')
plt.legend()
plt.show()
```

### 파이 차트

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False) # 그래프 시작 위치와 방향
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10] # explode: 중심에서 벗어난 정도

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

### Boxplot

```python
iris = sns.load_dataset('iris')
x = iris[iris.species == 'setosa'].sepal_length
y = iris[iris.species == 'setosa'].sepal_width
z = iris[iris.species == 'setosa'].petal_length
w = iris[iris.species == 'setosa'].petal_width
plt.boxplot([x, y, z, w], labels=['Sepal length', 'Sepal width', 'Petal length', 'Petal width'])
plt.title('Setosa')
plt.show()
```
