```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib
```

```python
driver = webdriver.Chrome(r'G:\\Workspace\\Data-analysis\\Data-analysis\\chromedriver.exe')

present_candi_text = []
for i in range(10):
    base_url = 'https://search.naver.com/search.naver'
    page_url = f'?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%EC%97%AC%EC%9E%90%EC%B9%9C%EA%B5%AC+%EC%84%A0%EB%AC%BC&c_id=&c_name=&sm=tab_pge&kin_start={1+10*i}&kin_age=0'
    driver.get(base_url+page_url)
    time.sleep(1)
    q_list = driver.find_elements_by_css_selector('.question_group')
    all_answers = ''
    for q in q_list:
        q.find_element_by_tag_name('a').click()
        time.sleep(2)
        driver.window_handles
        first_tab = driver.window_handles[0]
        last_tab = driver.window_handles[-1]
        driver.switch_to.window(window_name=last_tab)
        more = driver.find_element_by_css_selector('.answer-content__inner').find_element_by_id('nextPageButton')
        if int(driver.find_element_by_css_selector('._answerCount').text) <= 5:
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            a_list = soup.select('.answer-content__item')
            answers = ''
            for a in a_list:
                content = ''
                spans = a.select('.se-text-paragraph')
                for span in spans:
                    content += span.find('span', {'class': 'se-fs-'}).string
                answers += content
            driver.close()
            driver.switch_to.window(window_name=first_tab)
            all_answers += answers
            time.sleep(1)
        else:
            cur_cnt = int(driver.find_elements_by_css_selector('.c-button-more__count-number')[0].text)
            tot_cnt = int(driver.find_elements_by_css_selector('.c-button-more__count-number')[1].text)
            while (tot_cnt - cur_cnt) != (tot_cnt % cur_cnt):
                more.click()
                cur_cnt += 5
                time.sleep(1)
            if tot_cnt % 5 != 0:
                more.click()
                time.sleep(1)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            a_list = soup.select('.answer-content__item')
            answers = ''
            for a in a_list:
                content = ''
                spans = a.select('.se-text-paragraph')
                for span in spans:
                    content += span.find('span', {'class': 'se-fs-'}).string if span.find('span', {'class': 'se-fs-'}).string              else:
                answers += content
            driver.close()
            driver.switch_to.window(window_name=first_tab)
            all_answers += answers
            time.sleep(1)
    present_candi_text.append(all_answers)
```

```python
present_candi_text
f = open('./data/present.txt', 'w')
f.write(present_candi_text)
f.close()
```

```python
import nltk
from konlpy.tag import Twitter; t = Twitter()
```

```python
present_text = ''

for each_line in present_candi_text[:10000]:
    present_text = present_text + each_line
```

```python
tokens_ko = t.morphs(present_text)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name='여자친구 선물')
print(len(ko.tokens))
print(len(set(ko.tokens)))
```

```python
ko = nltk.Text(tokens_ko, name='여자친구 선물')
ko.vocab().most_common(100)
```

```python
stop_words = ['.', '이', '\u200b', ',', '을', '도', '에', '가', '으로', '는', '를', '!', '로', '의', '은', '수', '들', '요', '~', '한', '분', '나', '같아요', '\u200b\u200b', '좋은', '있는', '입니다', '하고', '것', '!!', '저', '?', '많이', '제', '거', '고', '하는', '적', '과', '^^', '..', '있습니다', '에서', '너무', '때', ')', '드립니다', '(', '할', '중', '향', '안녕하세요', '면', '드려요', '더', '인', '대', '다', '해', '등', '에게', '많은', '와', '요즘', '-', '합니다', '이나', '특별한', ':)', '정말', '4', '스', '까지', '잘', '"', "'", '/']

tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]

ko = nltk.Text(tokens_ko, name='여자친구 선물')
ko.vocab().most_common(50)
```
